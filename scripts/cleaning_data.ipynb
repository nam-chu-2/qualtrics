## Nearest airport city

Map each `airport_iata` to the city its airport serves. Keyed on IATA rather than the
airport name, since the names follow no consistent pattern ("Edmonton Intl" vs
"Sault Ste. Marie Airport" vs "Greater Moncton Romeo LeBlanc Intl").

```python
clean_path = "../canada_fsa_climate_footprint_clean.csv"
clean = pd.read_csv(clean_path)

# IATA -> city the airport serves. Where the airport's legal municipality differs from
# the city it serves, this takes the city served: YVR sits in Richmond, YYZ in Mississauga,
# and Bagotville is a borough of Saguenay.
AIRPORT_CITY = {
    "YAM": "Sault Ste. Marie",
    "YAT": "Attawapiskat",
    "YBC": "Baie-Comeau",
    "YBG": "Saguenay",
    "YBR": "Brandon",
    "YCB": "Cambridge Bay",
    "YCD": "Nanaimo",
    "YDF": "Deer Lake",
    "YEG": "Edmonton",
    "YFC": "Fredericton",
    "YGL": "Radisson",
    "YGR": "Îles-de-la-Madeleine",
    "YHO": "Hopedale",
    "YHZ": "Halifax",
    "YLW": "Kelowna",
    "YMM": "Fort McMurray",
    "YMX": "Mirabel",
    "YOW": "Ottawa",
    "YPL": "Pickle Lake",
    "YPR": "Prince Rupert",
    "YPX": "Puvirnituq",
    "YQB": "Quebec City",
    "YQL": "Lethbridge",
    "YQM": "Moncton",
    "YQR": "Regina",
    "YQT": "Thunder Bay",
    "YQU": "Grande Prairie",
    "YRB": "Resolute",
    "YSB": "Sudbury",
    "YTH": "Thompson",
    "YUL": "Montréal",
    "YVQ": "Norman Wells",
    "YVR": "Vancouver",
    "YWG": "Winnipeg",
    "YXE": "Saskatoon",
    "YXH": "Medicine Hat",
    "YXJ": "Fort St. John",
    "YXS": "Prince George",
    "YXU": "London",
    "YXY": "Whitehorse",
    "YYC": "Calgary",
    "YYD": "Smithers",
    "YYT": "St. John's",
    "YYZ": "Toronto",
    "YZF": "Yellowknife",
    "YZV": "Sept-Îles",
}

missing = set(clean["airport_iata"]) - set(AIRPORT_CITY)
assert not missing, f"unmapped IATA codes: {sorted(missing)}"

clean["nearest_airport_city"] = clean["airport_iata"].map(AIRPORT_CITY)
assert clean["nearest_airport_city"].notna().all()

clean.to_csv(clean_path, index=False)
clean[["fsa", "nearest_major_airport", "airport_iata", "nearest_airport_city"]].head()
```

## CO2 emissions cost

Price each FSA's five round-trip footprints under all six carbon tax regimes.

The tax values are **$/tonne** CO2 while the footprints are in **kg**, so each cost is
`(co2e_kg / 1000) * tax`, rounded to two significant figures. The CAD figures in the
parameters file already have the 1.4 exchange rate baked in — don't apply it again.

```python
from math import log10, floor


def round_sig(x, sig=2):
    if x == 0:
        return 0
    return round(x, sig - int(floor(log10(abs(x)))) - 1)


# Row 2 of the parameters file holds the rates; everything below it is reference notes.
taxes = pd.read_csv(
    "../footprint_parameters_June_16 pm(Tax Regimes).csv",
    encoding="utf-8-sig",
    nrows=1,
    usecols=range(6),
).iloc[0]

ROUTES = {
    "cdg": "co2e_cdg_kg",
    "lhr": "co2e_lhr_kg",
    "nrt": "co2e_nrt_kg",
    "set1": "co2e_set1_kg",
    "set2": "co2e_set2_kg",
}

co2 = clean[
    ["fsa", "name_/_area", "rovince_/_territory", "nearest_airport_city", *ROUTES.values()]
].copy()

for regime, rate in taxes.items():  # US_1..US_3 in USD, CAD_1..CAD_3 in CAD
    for route, kg_col in ROUTES.items():
        tonnes = clean[kg_col] / 1000
        co2[f"{regime.lower().replace('_', '')}_{route}"] = (tonnes * rate).map(round_sig)

co2.to_csv("../co2_emissions.csv", index=False)
print(taxes.to_dict())
co2.head()
```
